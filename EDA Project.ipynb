```python
#Data Cleaning and Preparation:
df.shape
```

```python
df.columns = df.columns.str.lower()
df
```

```python
df.isnull().sum()
```

```python
if 'price' in df.columns:
    df['price'] = df['price'].fillna(df['price'].mean())
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x='price', data=df)
plt.title("Boxplot of Price")
plt.xlabel("Price")
plt.ylabel("Listings")
plt.show()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound, upper_bound)
```

```python
df=df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
df
```

#Missing Values:
Missing values in numerical columns were replaced using the mean to preserve data distribution and avoid loss of information.

#Duplicate Records:
Duplicate rows were identified and removed to ensure data integrity and avoid repeated observations.

#Outliers:
Outliers in the availability_365 column were detected using boxplots and treated using the IQR method. Values outside the acceptable range were removed to improve analysis accuracy.

#Data Imbalance:
Categorical variables such as room_type were analyzed using count plots. The dataset showed slight imbalance, with certain room types appearing more frequently.
